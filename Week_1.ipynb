## L1.8 Data Types 2

```python
# boolean values
a = True #T & F should be capital
b = False
print(type(a))
print(type(b))
```

```python
# type conversion
a = int(5.7)
b = int('10')
print(a, type(a))
print(b, type(b))

# also if we take float(9), it'll give 9.0
```

```python
# in type conversion for boolean, only 0 and '' gives false -- rest all true
a = bool(10)
b = bool(0)
c = bool('India')
d = bool('0')
e = bool('') # boolean conversion of empty string is false

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
```
